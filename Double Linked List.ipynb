```python
class Node :
    def __init__ (self,data) :
        self.data = data
        self.next = None
        self.prev = None
        
    def setdata (self,data):
        self.data = data
    def getdata (self) :
        return self.data
    
    def setnext (self,n) :
        self.next = n
    def getnext (self) :
        return self.next
    
    def setprev (self,p):
        self.prev = p
    def getprev (self) :
        return self.prev
    
class doubleLinkedlist :
    def __init__ (self) :
        self.head = None
        
    def insert (self,new_data):
        new_node = Node(new_data)
        
        if (self.head == None) :
            self.head = new_node
        else :
            print("Data pertama sudah ada, gunakan metode insert lain !")
    
    def insertfirst (self,new_data):
        new_node = Node(new_data)
        last = self.head
        
        if (self.head == None) :
            self.head = new_node
        
        new_node.setnext(last)
        last.setprev(new_node)
        self.head = new_node
        new_node.setprev(self.head)   
    
    def insertafter (self,prev_data,new_data) :
        new_node = Node(new_data)
        next_node = prev_data.next
        
        new_node.setnext(prev_data.next)
        new_node.setprev(prev_data)
        next_node.setprev(new_node)
        prev_data.setnext(new_node)
    
    def insertlast(self,new_data) :
        new_node = Node(new_data)
        
        if self.head == None :
            self.head = new_node
            return
        last = self.head 
        while (last.next) :
            last = last.next
        last.next = new_node
        new_node.setprev(last)
    
    def remove (self,d) : 
        value = self.head
        self.d = d
        while (value) :
            if value.data == self.d :
                n = value.next
                p = value.prev
                
                n.setprev(p)
                p.setnext(n)
                
                return True
            else :
                value = value.next
    
    def view (self) :
        value = self.head
        
        while (value) :
            print(value.data)
            value = value.next
            
    def test (self) :
        last = self.head.getprev()
        
        print(last.data)
        
   
        
llist = doubleLinkedlist()

llist.insert(3)
llist.insertfirst(4)
llist.insertfirst(5)

llist.insertafter(llist.head,2)
llist.insertafter(llist.head.next,3)
llist.insertlast(7)

llist.view()
print("Data = ",llist.head.next.next.prev.data)
```
