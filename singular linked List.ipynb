```python
class Node :
    def __init__ (self,data) :
        self.data = data
        self.next = None
    
    def setnext (self,n) :
        self.next = n
    
    def getnext (self):
        return self.next

class singularLinkedlist :
    def __init__ (self):
        self.head = None
    
    def add (self,new_data):
        new_node = Node(new_data)
        
        if (self.head == None):
            self.head = new_node
            self.head.setnext(self.head)
        else :
            new_node.setnext(self.head.getnext())
            self.head.setnext(new_node)
    
    def remove (self,d) :
        print("\nafter remove ",d)
        cur = self.head
        prevnode = None
        
        while True :
            if cur.data == d :
                if prevnode is not None :
                    prevnode.setnext(cur.getnext())
                else :
                    while cur.getnext() != self.head :
                        cur = cur.getnext()
                    cur.setnext(self.head.getnext())
                    self.head = self.head.getnext()
                    
                    return True
                
            elif cur.getnext() == self.head :
                return False
            
            prevnode = cur
            cur = cur.getnext()
    
    def view (self) :
        cur = self.head
        
        print ("\nprint :")
        if (cur == None):
            print("Node Kosong !")
            return
        
        print(cur.data)
        
        while (cur.getnext() != self.head) :
            cur = cur.getnext()
            print(cur.data)

llist = singularLinkedlist()

llist.add(2)
llist.add(5)
llist.add(8)
llist.add(7)

llist.view()

llist.remove(2)

llist.view()
```
